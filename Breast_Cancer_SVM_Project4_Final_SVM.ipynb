# Convert diagnosis object type into int using encoding

```python

from sklearn.preprocessing import LabelEncoder
```

```python
le_diagnosis=LabelEncoder()
```

```python
df['diagnosis_new']=le_diagnosis.fit_transform(df['diagnosis'])
```

```python
df['diagnosis_new']
```

```python
df.shape
```

```python
#Check for null values
missing_data=df.isnull()
```

```python
missing_data
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("\n")
```

```python
#Replace Nan value
import numpy as np
df.replace("?",np.nan,inplace=True)
```

```python
#Defining X and Y 
```

```python
X = df.drop(['diagnosis'],axis=1)
X = np.asarray(X)
X[0:5]
```

```python
Y=df['diagnosis_new']
Y[0:5]
```

# Splliting Data

```python

X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  Y_train.shape)
print ('Test set:', X_test.shape,  Y_test.shape)
```

# Create Model

```python
#Create Model
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, Y_train) 
```

```python
X_train[0:5]
```

```python
X_train.shape
```

```python
Y_test.shape
```

```python
# Acuatl values
Y_test[0:5]
```

```python
#Predicted value
yhat = clf.predict(X_test)
yhat [0:5]
```

# Accuracy evaluation of model

```python
#Accuracy evaluation of model
print(classification_report(Y_test,yhat))
```

```python
# to increase accuracy of model scaling
min_train =X_train.min()
range_train =(X_train - min_train).max()
X_train_scaled =(X_train-min_train)/range_train
```

```python
min_test =X_test.min()
range_test =(X_test - min_test).max()
X_test_scaled =(X_test-min_test)/range_test
```

```python
clf.fit(X_train_scaled,Y_train)
```

```python
Y_predict =clf.predict(X_test_scaled)
```

```python
#Accuracy of model after scaling
print(classification_report(Y_test,Y_predict))
```
